### Street with highest and lowest and highest number of bike riders for different years

```python
bike_traffic = df_bike_2020.drop('Month', axis=1).sum()

# Find the street with the highest total bicycle traffic
street_with_most_bikers = bike_traffic.idxmax()
street_with_least_bikers = bike_traffic.idxmin()


# Print the street with the most number of bikers
print("Street with the most number of bikers in 2020:", street_with_most_bikers)
# Print the street with the least traveled
print("Street with the least traveled in 2020:", street_with_least_bikers)
```

### Bar plot of Total Bicycle Traffic per month for year 2020, 2021, 2022, for street Venloer Straße

```python
#Bar plot of Total Bicycle Traffic per month for year 2020, 2021, 2022, for street Venloer Straße
import matplotlib.pyplot as plt

monthly_traffic_2020 = df_bike_2020.groupby('Month').sum()
monthly_traffic_2021 = df_bike_2021.groupby('Month').sum()
monthly_traffic_2022 = df_bike_2022.groupby('Month').sum()

plt.figure(figsize=(10, 6))
plt.bar(monthly_traffic_2020.index, monthly_traffic_2020['Venloer Straße'], label='Venloer Straße 2020')
plt.bar(monthly_traffic_2021.index, monthly_traffic_2021['Venloer Straße'], label='Venloer Straße 2021')
plt.bar(monthly_traffic_2022.index, monthly_traffic_2022['Venloer Straße'], label='Venloer Straße 2022')
plt.xlabel('Month')
plt.ylabel('Total Bicycle Traffic')
plt.title('Total Bicycle Traffic per Month')
plt.xticks(rotation=45)
plt.show()
```
